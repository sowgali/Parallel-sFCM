```python
import numpy as np
import PIL
from PIL import Image, ImageOps
img = Image.open("test.jpg")
```

```python
gr_img = ImageOps.grayscale(img)
new_arr = np.array(gr_img)
new_arr.shape
```

```python
gr_img
```

```python
new_arr = new_arr-new_arr.mean()/new_arr.std() 
new_arr
```

```python
with open("sample.txt", "w") as f:
    for row in new_arr:
        for col in row:
            f.write(str(float(col)))
            f.write(" ")
        f.write("\n")
```
